**Theory Questions: **

1. What are data structures, and why are they important?
->Data structures are ways of organizing and storing data to enable efficient access and modification. They are important because they help in writing optimized code, improving performance, and solving complex problems effectively.

2. Explain the difference between mutable and immutable data types with examples?
-> Mutable data types can be changed after creation, like lists ([1, 2, 3] can become [1, 2, 4]). Immutable data types cannot be changed after creation, like tuples ((1, 2, 3)) or strings ("hello")

3.  What are the main differences between lists and tuples in Python?
-> The main differences between lists and tuples in Python are:
Mutability: Lists are mutable (can be changed), while tuples are immutable (cannot be changed).
Syntax: Lists use square brackets [ ], tuples use parentheses ( ).
Performance: Tuples are generally faster and use less memory than lists.

4.  Describe how dictionaries store data?
-> Dictionaries in Python store data as key-value pairs. Each key is unique and maps to a value, allowing fast access to data using the key. Internally, dictionaries use a hash table to quickly find values.

5. Why might you use a set instead of a list in Python?
-> We might use a set instead of a list when you need to store unique elements and perform fast membership tests. Sets automatically remove duplicates and offer faster lookup compared to lists.

6. What is a string in Python, and how is it different from a list?
-> A string in Python is a sequence of characters enclosed in quotes, like "hello". Unlike lists, strings are immutable (cannot be changed) and specifically hold text, while lists can hold any data type and are mutable.

7.  How do tuples ensure data integrity in Python?
-> Tuples ensure data integrity in Python by being immutable, meaning their values cannot be changed after creation. This makes them ideal for storing fixed, constant data and helps prevent accidental modifications.

8. What is a hash table, and how does it relate to dictionaries in Python?
-> A hash table is a data structure that maps keys to values using a hash function to compute an index. In Python, dictionaries use hash tables internally to store key-value pairs, allowing for fast access, insertion, and deletion

9.  Can lists contain different data types in Python?
-> Yes, lists in Python can contain different data types, such as integers, strings, floats, or even other lists. Example: [1, "hello", 3.14, [2, 4]].

10. Explain why strings are immutable in Python?
-> Strings are immutable in Python to ensure data integrity, security, and performance. This allows strings to be safely shared, cached, and used as dictionary keys without risk of accidental changes.

11. What advantages do dictionaries offer over lists for certain tasks ?
-> Dictionaries offer faster data lookup using keys and allow key-value pair storage, making them ideal for tasks like counting items, storing configurations, or mapping relationships. Unlike lists, they provide constant-time access to elements by key.

12. Describe a scenario where using a tuple would be preferable over a list?
-> Using a tuple is preferable when you need to store a fixed set of values that shouldn’t change, like coordinates (x, y) or dates. Tuples are also more memory-efficient and can be used as keys in dictionaries.

13. How do sets handle duplicate values in Python?
-> Sets in Python automatically remove duplicate values. When you add duplicates to a set, only one instance is kept, ensuring all elements are unique.

14. How does the “in” keyword work differently for lists and dictionaries?
-> In lists, the in keyword checks if a value exists in the list. In dictionaries, it checks if a key exists, not the value.
Example:
5 in [1, 2, 5] → True
"name" in {"name": "Sahithi"} → True

15. Can you modify the elements of a tuple? Explain why or why not?
-> No, you cannot modify the elements of a tuple because tuples are immutable. Once created, their content cannot be changed, which helps ensure data stays constant and safe from accidental updates.

16. What is a nested dictionary, and give an example of its use case?
-> A nested dictionary is a dictionary within another dictionary, used to represent complex, structured data.

Example use case: Storing student records:
students = {
    "101": {"name": "Alice", "grade": "A"},
    "102": {"name": "Bob", "grade": "B"}
}

17. Describe the time complexity of accessing elements in a dictionary?
-> Accessing elements in a dictionary has an average-case time complexity of O(1) due to its use of a hash table. This allows for fast lookups by key, regardless of the dictionary size.

18. In what situations are lists preferred over dictionaries?
-> Lists are preferred over dictionaries when you need to store ordered sequences of items, especially when position matters or when duplicate values are needed. They are ideal for tasks like iterating, sorting, or storing simple collections.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
-> Dictionaries were considered unordered before Python 3.7, meaning the insertion order wasn't preserved. Since Python 3.7, they maintain insertion order, but they’re still designed for key-based access, not position. This affects data retrieval by focusing on keys, not order.

20. Explain the difference between a list and a dictionary in terms of data retrieval?
-> In a list, data is retrieved using index positions (e.g., my_list[0]), while in a dictionary, data is retrieved using keys (e.g., my_dict["name"]). Lists are ideal for ordered data, dictionaries for labeled data.
